Droping Runtime, Seasons, Imdb_id, Id , Tmdb_score\r Columns

```python
data.columns = data.columns.str.strip()
cols = ['runtime', 'seasons', 'imdb_id', 'id', 'tmdb_score']
data.drop(cols, axis=1, inplace=True)
data.columns
```

```python
data.head()
```

Checking for Null Values

```python
data.isnull().sum()
```

Droping Tuple That Has NAN as the Title

```python
data = data.dropna(subset=['title'])
data.isnull().sum()
```

Handling Null Values

```python
data['description'].fillna('unknown',inplace=True)
data['age_certification'].fillna('unknown',inplace=True)
data['imdb_score'].fillna(data['imdb_score'].mean(),inplace=True)
data['imdb_votes'].fillna(data['imdb_votes'].mean(),inplace=True)
data['tmdb_popularity'].fillna(data['tmdb_popularity'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
print('handling null values')
data
```

Removing Square Brackets and Single Quotes From Genres and Production Countries

```python
data['genres'] = data['genres'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
import ast
data['genres'] = data['genres'].apply(lambda x: ', '.join(ast.literal_eval(x)) if isinstance(x, str) else x)
data
```

```python
data['production_countries'] = data['production_countries'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
import ast
data['production_countries'] = data['production_countries'].apply(lambda x: ', '.join(ast.literal_eval(x)) if isinstance(x, str) else x)
data
```

Handling Blank Spaces in Production Countries

```python
data['production_countries'] = data['production_countries'].replace('', 'Unknown')
data
```

Analysis for RELEASE YEARS

CATEGORIZING release_year into ----> old, classic, recent, latest

```python
data.head()
```

```python
data['release_year'].describe()
```

```python
def categorize_col(data, col, labels):
    edges = [data[col].describe()['min'],
             data[col].describe()['25%'], 
             data[col].describe()['50%'],
             data[col].describe()['75%'], 
             data[col].describe()['max']]
    data[col] = pd.cut(data[col], edges, labels=labels, duplicates='drop', include_lowest=True)
    return data

labels = ['old', 'classic', 'recent', 'latest']
data = categorize_col(data, 'release_year', labels)
print(data['release_year'].unique())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='release_year', order=data['release_year'].cat.categories ,palette='mako')
plt.title('Distribution of Release Year Categories')
plt.xlabel('Release Year Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

Analysis of Tmdb Popularity

```python
data = data.dropna(subset=['tmdb_popularity'])
bins = [0, 10, 25, 50, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']
data['popularity_category'] = pd.cut(data['tmdb_popularity'], bins=bins, labels=labels, include_lowest=True)
popularity_counts = data['popularity_category'].value_counts().reindex(labels)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x=popularity_counts.index, y=popularity_counts.values, palette='crest')
plt.title('TMDb Popularity Distribution (Your Dataset)', fontsize=14)
plt.xlabel('Popularity Category')
plt.ylabel('Number of Titles')
for i, count in enumerate(popularity_counts.values):
    plt.text(i, count + 20, str(int(count)), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
